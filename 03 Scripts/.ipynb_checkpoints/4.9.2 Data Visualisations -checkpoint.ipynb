## 1. Importing libraries and data sets

## Barchart - orders day of the week

```python
#count rows for each day of the week
ords_prods['orders_day_of_week'].value_counts()
```

```python
#bar chart on orders_day_of_week value counts to see the distribution. Sorted by largest to smallest

plt.title('Orders Day of the Week Distribution', fontsize=13)
bar_odow = ords_prods['orders_day_of_week'].value_counts().sort_index().plot.bar(color=['green'], edgecolor='black', xlabel='orders day of the week', ylabel='frequency')

#formats y-axis as M (millions)

bar_odow.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x,pos: format(x/1000000,'1.0f')+'M'))
```

### Export Barchart orders day of the week distribution

```python
bar_odow.figure.savefig(os.path.join(r'/Users/katja/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/bar_orders_dow.png'))
```

```python
#delete df to save space 

del bar_odow

gc.collect()
```

###  Barchart orders day of the week distribution 2  incl. weekdays

```python
#bar chart on orders_day_of_week value counts to see the distribution. Sorted by largest to smallest

plt.title('Orders by Day of the Week', fontsize=13)
bar = ords_prods['orders_day_of_week'].value_counts().sort_index().plot.bar(color=['green'], edgecolor='black', xlabel='orders day of the week', ylabel='frequency')

# Set custom axis labels
bar.set_xlabel("")  # Set an empty string as the x-axis label

# Customize the x-axis ticks and labels
custom_xticks = range(7)  # Define your custom tick positions (e.g., for days of the week)
custom_xlabels = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]  # Define your custom tick labels

bar.set_xticks(custom_xticks)
bar.set_xticklabels(custom_xlabels,rotation=360)

#formats y-axis as M (millions)

bar.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x,pos: format(x/1000000,'1.0f')+'M'))
```

### Export Barchart orders day of the week distribution 2 (incl. weekdays)

```python
bar.figure.savefig(os.path.join(r'/Users/katja/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/bar_orders_2_incl_weekdays2.png'))
```

```python
#delete df to save space 

del bar

gc.collect()
```

## Linechart - prices vs orders day of week

### create sampling data by spliting dataframe 70/30

```python
# spliting dataframes into 70/30 by assigning 'True' to random number less than 0.7 and 'False' to more

#create the random seed so its replicable
np.random.seed(4)

#create a list of (True, False) values based on random assignment of values between 0-1 and is less than 0.7
dev = np.random.rand(len(ords_prods)) <= 0.7
```

```python
# Store 70% of data sample in big dataframe (<= 0.7)
big = ords_prods[dev]
```

```python
# Stores 30% of data sample in small dataframe (> 0.7)
small = ords_prods[~dev]
```

```python
# Checking the length of the dataframes
big_len = len(big)
small_len = len(small)

big_len + small_len
```

## Analyse age vs spending power (income) via scatterplot

```python
# Extract age and income into a separate dataframe
df_age_income = small [['age', 'income']]
df_age_income.head()
```

```python
# Create a scatterplot to see the correlation between age and income
scatter_age_income = sns.scatterplot(x='age', y='income',color='green', data=df_age_income)
plt.title('Age vs Income (spending power)')
scatter_age_income.set(xlabel='age', ylabel='income')
```

### Export Scatterplot Age vs Income (spending power)

```python
scatter_age_income.figure.savefig(os.path.join(r'/Users/katja/Documents/Instacart Basket Analysis/04 Analysis/Visualizations/scatter_age_income.png'))
```

```python
#delete df to save space 

del scatter_age_income
del df_age_income

gc.collect()
```

#### Observations:

#### Customer between 18 to 40 years old have a lower spending power compared to above the above 40 years old. 
#### The majority of customer between 18 and 40 years old have income below 400,000 USD. Most customers have an income less than 200,000 USD.
#### Where as customers above 40 years old surpass the income sealing of 400,000 USD and have an income up to 600,000 USD, meaning the spending power increases above the age of 40 years old.

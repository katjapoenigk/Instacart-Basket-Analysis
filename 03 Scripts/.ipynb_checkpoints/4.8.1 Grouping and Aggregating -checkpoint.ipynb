## Create a subset & use groupby

```python
#creating a subset

df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
df.head(10)
```

## Grouping data with panda

```python
#grouping product_name

df.groupby('product_name')
```

```python
#via agg 'mean' of 'order_number' which previously was renamed as 'clients_order_number'

df.groupby('department_id').agg({'clients_order_number': ['mean']})
```

```python
#calculating mean

df.groupby('department_id')['clients_order_number'].mean()
```

```python
#calculating mean, and min max, performing multiple aggregations

df.groupby('department_id').agg({'clients_order_number': ['mean', 'min', 'max']})
```

## Creating a loyalty flag

```python
#creating max order column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['clients_order_number'].transform(np.max)
```

```python
#turning off max rows

pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(15)
```

```python
#Loyal Customer LOC tag

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Regular Customer LOC tag

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#New Customer LOC tag

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#value count loyalty customer

ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```
